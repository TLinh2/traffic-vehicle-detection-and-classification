```python
from pathlib import Path

ROOT = Path("/home/team_thuctap/ltlinh/new_ndanh/experiments/v9(100epochs)_debug_bckgnd_yolo11m_2015-12-19/error_analysis/visuals")
ORDER = ["false_positive", "false_negative", "wrong_class"]
IMG_EXTS = {".jpg", ".png", ".jpeg"}

seen = set()
result = []

for folder in ORDER:
    folder_path = ROOT / folder
    if not folder_path.exists():
        continue

    for p in sorted(folder_path.iterdir()):
        if p.suffix.lower() not in IMG_EXTS:
            continue

        name = p.stem
        if name in seen:
            continue

        seen.add(name)
        result.append(name)

# ---- save ----
with open("error_images.txt", "w") as f:
    for x in result:
        f.write(x + "\n")

print(f"[DONE] Collected {len(result)} unique image names")
```
